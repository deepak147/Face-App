```python
from glob import glob
paths = glob('./UTKFace/*.jpg')
paths
```

```python
for i, path in enumerate(paths):
    img = cv2.imread(path)
    try:
        rimg = cv2.resize(img, (224, 224), cv2.INTER_CUBIC)
        print(path.split('_')[1])
        if path.split('_')[1] == '0':
            cv2.imwrite('./male/male_{}.jpg'.format(i),rimg)
            print('saved successfully male_{}.jpg'.format(i))
        if path.split('_')[1] == '1':
            cv2.imwrite('./female/female_{}.jpg'.format(i),rimg)
            print('saved successfully female_{}.jpg'.format(i))
    except:
        print('not successful')
```
